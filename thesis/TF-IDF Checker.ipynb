```python
import tkinter as tk
from tkinter import Text, Button, Label, messagebox
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib
```

```python
data = pd.read_csv("Combined Total - Preprocessed (Labeled).csv", delimiter=',' )
label_counts = data['label'].value_counts()
print(label_counts)
data
```

```python
min_df_value = 3
tfidf_vectorizer = TfidfVectorizer(min_df=min_df_value, ngram_range=(2,2), smooth_idf=False)
X_tfidf = tfidf_vectorizer.fit_transform(data['text'])
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names())
tfidf_df
```

```python
sample_text = "wake up you stoopid dyke"
sample_tfidf = tfidf_vectorizer.transform([sample_text])

print("TF-IDF values for the sample text for Overall")
for term, tf_idf_value in zip(tfidf_vectorizer.get_feature_names(), sample_tfidf.toarray()[0]):
    if tf_idf_value != 0.0:
        # Print the TF-IDF value
        print(term, "=", tf_idf_value)
```

```python
tfidf_with_label = pd.concat([tfidf_df, data['label']], axis=1)
tfidf_with_label
```

```python
class1_documents = tfidf_with_label[tfidf_with_label['label'] == 1]
class1_tfidf_sum = class1_documents.sum(axis=0)
class1_tfidf_sum = class1_tfidf_sum.drop(tfidf_with_label.columns[-1])  

print("Sum of TF-IDF for terms in Hate documents:")
for term, tfidf_sum in zip(class1_tfidf_sum.index, class1_tfidf_sum.values):
  print(f"- {term}: {tfidf_sum:.2f}")
```

```python
class0_documents = tfidf_with_label[tfidf_with_label['label'] == 0]
class0_tfidf_sum = class0_documents.sum(axis=0)
class0_tfidf_sum = class0_tfidf_sum.drop(tfidf_with_label.columns[-1])  

print("Sum of TF-IDF for terms in Not Hate documents:")
for term, tfidf_sum in zip(class0_tfidf_sum.index, class0_tfidf_sum.values):
  print(f"- {term}: {tfidf_sum:.2f}")
```

```python
#FOR HATE
sample_terms = [term for term, tf_idf_value in zip(tfidf_vectorizer.get_feature_names(), sample_tfidf.toarray()[0]) if tf_idf_value != 0.0]

matching_terms_hate = []
for term in sample_terms:
  if term in class1_tfidf_sum.index:
    matching_terms_hate.append(term)
    
if matching_terms_hate:
  print("Matching terms in Hate:")
  for term in matching_terms_hate:
    tfidf_sum = class1_tfidf_sum.loc[term]
    print(f"{term} = {tfidf_sum:.2f}")
else:
  print("No matching terms found.")
```

```python
#FOR NOT HATE
sample_terms = [term for term, tf_idf_value in zip(tfidf_vectorizer.get_feature_names(), sample_tfidf.toarray()[0]) if tf_idf_value != 0.0]

matching_terms_nothate = []
for term in sample_terms:
  if term in class0_tfidf_sum.index:
    matching_terms_nothate.append(term)
    
if matching_terms_nothate:
  print("Matching terms in Not Hate:")
  for term in matching_terms_nothate:
    tfidf_sum = class0_tfidf_sum.loc[term]
    print(f"{term} = {tfidf_sum:.2f}")
else:
  print("No matching terms found.")
```
