```python
import tkinter as tk
from tkinter import Text, Button, Label, messagebox
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib
```

```python
data = pd.read_csv("Combined Total - Preprocessed (Labeled).csv", delimiter=',' )
label_counts = data['label'].value_counts()
print(label_counts)
data
```

```python
min_df_value = 3 # set to 2: sets the minimum frequency required for a term to be considered
tfidf_vectorizer = TfidfVectorizer(min_df=min_df_value, ngram_range=(2,2), smooth_idf=False) # 2,2= Bigram
X_tfidf = tfidf_vectorizer.fit_transform(data['text'])
tfidf_df = pd.DataFrame(X_tfidf.toarray(), columns=tfidf_vectorizer.get_feature_names())
tfidf_df
```

```python
# Concatenate TF-IDF DataFrame with the label column
tfidf_with_label = pd.concat([tfidf_df, data['label']], axis=1)
top_tfidf_all_labels = tfidf_with_label.drop('label', axis=1).sum().sort_values(ascending=False).head(10)

# Filter TF-IDF DataFrame based on label value
top_tfidf_label_0 = tfidf_with_label[tfidf_with_label['label'] == 0].sum().sort_values(ascending=False).head(10)

# Print the top TF-IDF values for each label
print("Top TF-IDF values for label 0:")
print(top_tfidf_label_0)
```

```python
# Define a sample text/document
sample_text = "Aahon sa aba eh love me hate me sorry na commercial ni"

# Transform the sample text/document using the existing TfidfVectorizer instance
sample_tfidf = tfidf_vectorizer.transform([sample_text])

# Print the TF-IDF values for the sample text/document without considering labels
print("TF-IDF values for the sample text for Overall")
for term, tf_idf_value in zip(tfidf_vectorizer.get_feature_names(), sample_tfidf.toarray()[0]):
    if tf_idf_value != 0.0:
        # Print the TF-IDF value
        print(term, "=", tf_idf_value)
```

```python
# Get the indices of documents where the label is 0 and 1
label_0_indices = tfidf_with_label.index[tfidf_with_label['label'] == 0].tolist()
label_1_indices = tfidf_with_label.index[tfidf_with_label['label'] == 1].tolist()

# Filter the TF-IDF DataFrame to include only rows where the label is 0 and 1
tfidf_label_0 = tfidf_df.iloc[label_0_indices]
tfidf_label_1 = tfidf_df.iloc[label_1_indices]

# Filter the TF-IDF DataFrames to include only columns (features) present in each label subset
tfidf_label_0 = tfidf_label_0[tfidf_label_0.columns.intersection(tfidf_label_0.columns)]
tfidf_label_1 = tfidf_label_1[tfidf_label_1.columns.intersection(tfidf_label_1.columns)]

# Print the shape of the filtered DataFrames
print("Shape of TF-IDF subset for label 0:", tfidf_label_0.shape)
print("Shape of TF-IDF subset for label 1:", tfidf_label_1.shape)
```

```python
# Print the TF-IDF values for the sample text/document considering only documents labeled as 0
print("\nTF-IDF values for the sample text for Not Hate")
for term in tfidf_vectorizer.get_feature_names():
    # Check if the term exists in the vocabulary learned by the TfidfVectorizer
    if term in tfidf_vectorizer.vocabulary_:
        # Get the index of the term in the vocabulary
        term_index = tfidf_vectorizer.vocabulary_[term]
        
        # Access the TF-IDF value of the term in the sample text/document
        tf_idf_value_label_0 = sample_tfidf[0, term_index]
        
        # Check if the TF-IDF value is not equal to 0.0 and if the term exists in the TF-IDF values where the label is 0
        if tf_idf_value_label_0 != 0.0 and term in tfidf_label_0.columns:
            # Print the term and its TF-IDF value
            print(term, "=", tf_idf_value_label_0)
            
            
print("\nTF-IDF values for the sample text for Hate")
for term in tfidf_vectorizer.get_feature_names():
    # Check if the term exists in the vocabulary learned by the TfidfVectorizer
    if term in tfidf_vectorizer.vocabulary_:
        # Get the index of the term in the vocabulary
        term_index = tfidf_vectorizer.vocabulary_[term]
        
        # Access the TF-IDF value of the term in the sample text/document
        tf_idf_value_label_1 = sample_tfidf[0, term_index]
        
        # Check if the TF-IDF value is not equal to 0.0 and if the term exists in the TF-IDF values where the label is 1
        if tf_idf_value_label_1 != 0.0 and term in tfidf_label_1.columns:
            # Print the term and its TF-IDF value
            print(term, "=", tf_idf_value_label_1)
```
